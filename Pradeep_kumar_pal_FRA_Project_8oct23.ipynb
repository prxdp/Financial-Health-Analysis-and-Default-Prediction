```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
X = df_imputed.drop('Default', axis = 1)
calc_vif(X).sort_values(by = 'VIF', ascending = False)
```

```python
X = X.drop('_Net_profit_before_tax_to_Paid_in_capital', axis = 1)
calc_vif(X).sort_values(by = 'VIF', ascending = False)
```

```python
X = X.drop('_Quick_Assets_to_Current_Liability', axis = 1)
calc_vif(X).sort_values(by = 'VIF', ascending = False)
```

```python
X = X.drop('_CFO_to_Assets', axis = 1)
calc_vif(X).sort_values(by = 'VIF', ascending = False)
```

```python
dependent_variable = 'Default'

# Initial list of independent variables
independent_variables = [col for col in X.columns if col != dependent_variable]

while True:
    # Fit the logistic regression model with the current set of independent variables
    formula = f"{dependent_variable} ~ {' + '.join(independent_variables)}"
    model = SM.logit(formula=formula, data=Default_train).fit()

    # Get the p-values of the coefficients
    p_values = model.pvalues

    # Find the variable with the highest p-value
    max_p_value = p_values.drop('Intercept').max()
    variable_to_remove = p_values.idxmax()

    # Check if the highest p-value exceeds the threshold (e.g., 0.05)
    if max_p_value > 0.05:
        # Remove the variable with the highest p-value
        independent_variables.remove(variable_to_remove)
        print(f"Removed variable: {variable_to_remove} (p-value={max_p_value})")
    else:
        # If no variable exceeds the threshold, exit the loop
        break

# Display the final selected variables and model summary
final_formula = f"{dependent_variable} ~ {' + '.join(independent_variables)}"
final_model = SM.logit(formula=final_formula, data=Default_train).fit()
print("Final selected variables:")
print(independent_variables)
print("Final Model Summary:")
print(final_model.summary())
```
